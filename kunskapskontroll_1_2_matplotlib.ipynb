### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

In Matplotlib:

Figure: is like the canvas for your plot.
Axes: are individual plots or subplots on the canvas.
Axis: represents the scale of each subplot.
Artist: is any element like lines, text, or shapes on the plot.
These components work together to create and customize data visualizations in Matplotlib.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# OOs styel

# Generate x values from -4 to 4
x = np.linspace(-4, 4, 100)

# Calculate y values (y = x^2)
y = x**2

# Create a Figure and Axes objects
fig, ax = plt.subplots()

# Plot the data using the Axes object
ax.plot(x, y, label='y = x^2')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('OOs styel y = x^2')

# Display the legend
ax.legend()

# Show the plot
plt.show()
```
